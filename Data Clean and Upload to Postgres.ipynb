# All Crime Data

## Aim: visualise and predict time series data from WA Police. 

I created 17 csv files from a large Excel workbook available on the WA police website. The data shows occurances of 72 types of crime for month by month from Jan 2007 to December 2020.

```python
import pandas as pd
```

```python
# Create a reference to the fremantle CSV and import it into a Pandas DataFrame as a test
csv_path_1 = "CSV_crime/fremantle_crime.csv"
fremantle_df = pd.read_csv(csv_path_1)
fremantle_df.head()
```

```python
# Create a reference to all CSV files and import them into Pandas DataFrames
csv_path_2 = "CSV_crime/cannington_crime.csv"
cannington_df = pd.read_csv(csv_path_2)

csv_path_3 = "CSV_crime/goldfieldsesperance_crime.csv"
goldfieldsesperance_df = pd.read_csv(csv_path_3)

csv_path_4 = "CSV_crime/greatsouthern_crime.csv"
greatsouthern_df = pd.read_csv(csv_path_4)

csv_path_5 = "CSV_crime/joondalup_crime.csv"
joondalup_df = pd.read_csv(csv_path_5)

csv_path_6 = "CSV_crime/kimberley_crime.csv"
kimberley_df = pd.read_csv(csv_path_6)

csv_path_7 = "CSV_crime/mandurah_crime.csv"
mandurah_df = pd.read_csv(csv_path_7)

csv_path_8 = "CSV_crime/metropolitanregion_crime.csv"
metropolitanregion_df = pd.read_csv(csv_path_8)

csv_path_9 = "CSV_crime/midland_crime.csv"
midland_df = pd.read_csv(csv_path_9)

csv_path_10 = "CSV_crime/midwestgascoyne_crime.csv"
midwestgascoyne_df = pd.read_csv(csv_path_10)

csv_path_11 = "CSV_crime/mirrabooka_crime.csv"
mirrabooka_df = pd.read_csv(csv_path_11)

csv_path_12 = "CSV_crime/perth_crime.csv"
perth_df = pd.read_csv(csv_path_12)

csv_path_13 = "CSV_crime/pilbara_crime.csv"
pilbara_df = pd.read_csv(csv_path_13)

csv_path_14 = "CSV_crime/regionalwa_crime.csv"
regionalwa_df = pd.read_csv(csv_path_14)

csv_path_15 = "CSV_crime/southwest_crime.csv"
southwest_df = pd.read_csv(csv_path_15)

csv_path_16 = "CSV_crime/westernaustralia_crime.csv"
westernaustralia_df = pd.read_csv(csv_path_16)

csv_path_17 = "CSV_crime/wheatbelt_crime.csv"
wheatbelt_df = pd.read_csv(csv_path_17)
```

ðŸ¤“ With 17 files, it was necessary to concatanate them all to create one huge CSV file.

```python
#Concat two dataframes as a test
merge_df1 = pd.concat([fremantle_df, cannington_df], ignore_index=True, sort=False)
merge_df1.head()
```

```python
#Concat all df into one large df
merge_df2 = pd.concat([goldfieldsesperance_df, greatsouthern_df], ignore_index=True, sort=False)

merge_df3 = pd.concat([joondalup_df, kimberley_df], ignore_index=True, sort=False)

merge_df4 = pd.concat([mandurah_df, metropolitanregion_df], ignore_index=True, sort=False)

merge_df5 = pd.concat([midland_df, midwestgascoyne_df], ignore_index=True, sort=False)

merge_df6 = pd.concat([mirrabooka_df, perth_df], ignore_index=True, sort=False)

merge_df7 = pd.concat([pilbara_df, regionalwa_df], ignore_index=True, sort=False)

merge_df8 = pd.concat([southwest_df, westernaustralia_df], ignore_index=True, sort=False)

merge_df9 = pd.concat([wheatbelt_df, merge_df1], ignore_index=True, sort=False)

merge_df10 = pd.concat([merge_df2, merge_df3], ignore_index=True, sort=False)

merge_df11 = pd.concat([merge_df4, merge_df5], ignore_index=True, sort=False)

merge_df12 = pd.concat([merge_df6, merge_df7], ignore_index=True, sort=False)

merge_df13 = pd.concat([merge_df8, merge_df9], ignore_index=True, sort=False)

merge_df14 = pd.concat([merge_df10, merge_df11], ignore_index=True, sort=False)

merge_df15 = pd.concat([merge_df12, merge_df13], ignore_index=True, sort=False)

datamerged_df = pd.concat([merge_df14, merge_df15], ignore_index=True, sort=False)
```

# Upload the dataframe to Postgres

## I used sqlalchemy to load the data in a database (crime_db)

```python
#Import dependencies for creating a connection to postgres
from sqlalchemy import create_engine
import psycopg2
```

```python
#Load into postgres database (crime_db)
engine = create_engine(f'postgresql://postgres:postgres@localhost:5432/crime_db')
connection = engine.connect()
```

## Upload the data to a table called 'crime' within the crime_db

```python
#Create a 'crime table in postgres db
crime_df.to_sql('crime', connection, if_exists='replace')
```
